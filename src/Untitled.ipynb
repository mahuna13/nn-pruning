```python
def plot_with_color_bar(toPlot):
    plt.figure()
    ax = plt.gca()
    im = ax.imshow(toPlot)

    divider = make_axes_locatable(ax)
    cax = divider.append_axes("right", size="5%", pad=0.05)

    plt.colorbar(im, cax=cax)
plot_with_color_bar(net.weights[0])
```
